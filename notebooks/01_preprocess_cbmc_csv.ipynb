```python
import pandas as pd
import scanpy as sc
import numpy as np
```

```python
# Load the CSV into a DataFrame
df = pd.read_csv("../data/raw/cbmc_raw/GSE100866_CBMC_8K_13AB_10X-RNA_umi.csv.gz", index_col=0)
df
```

```python
# Transpose it (genes Ã— cells â†’ cells Ã— genes)
adata = sc.AnnData(df.T)
```

```python
# Basic filtering
sc.pp.filter_cells(adata, min_genes=200)
sc.pp.filter_genes(adata, min_cells=3)
```

```python
# Optional QC filters # NOT RUN
adata = adata[adata.obs.n_genes_by_counts < 2500, :]
adata = adata[adata.obs.pct_counts_mt < 5, :]
```

```python
# Normalize and log transform
sc.pp.normalize_total(adata, target_sum=1e4)
sc.pp.log1p(adata)
```

```python
# Select highly variable genes (optional but helpful for deep models)
sc.pp.highly_variable_genes(adata, min_mean=0.0125, max_mean=3, min_disp=0.5)
adata = adata[:, adata.var.highly_variable]
```

```python
# Scale
sc.pp.scale(adata, max_value=10)
```

```python
# Save final AnnData object
adata.write("../data/processed/cbmc.h5ad")
print("ðŸ˜Preprocessing complete. Shape of Adata:", adata.shape)
```
